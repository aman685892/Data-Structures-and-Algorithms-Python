```python
def addition(a,b):
  return a+b

addition(4,5)
```

```python
## using lambda function as anonymous function(anobymous function is a function without a name)
addition=lambda a,b:a+b  ## here addition is a variable that takes two parameters as a,b
print(addition(4,5))
print(type(addition))
```

```python
def even(num):
  if num%2==0:
    return True
even(34)
```

```python
even1=lambda num:  num%2==0
print(even1(56))
```

```python
## dealing with multiple features in functions
def addition(x,y,z):
  return x+y+z
addition(3,4,5)
```

```python
## using lambda as anonymous function
addition=lambda x,y,z:x+y+z
print(addition(4,5,6))
```

```python
## map()-applies function to every items in the given list
numbers=[1,2,3,4,5]
def square(numbers):
  return numbers**2
square(4)
```

```python
square=lambda x:x**2
print(square(9))
```

```python
list(map(lambda x:x**2,numbers))
```
