```python
def square(x): ## here i have defined the function named as square
  return x*x
```

```python
numbers=[1,2,3,4]
list(map(square,numbers))  ## here map function map between function named square and list as numbers
```

```python
## Lambda function with map()
numbers=[1,2,3,4]
list(map(lambda x:x**2,numbers))
```

```python
## mapping multiple iterables
num1=[1,2,3,4]
num2=[5,6,7,8]
list(map(lambda x,y:x+y,num1,num2))  ## Lambda takes only ONE EXPRESSION, not one parameter,it can takes multiple paramters
```

```python
## map( ) to covert a list of strings to integers
str_numbers=['1','2','3','4','5']
list(map(int,str_numbers))
```

```python
## inbulit function woth maps()
words=["maan","raju","papaya","apple"]
list(map(str.upper,words))## mapping a function with iterables
```

```python
def get_name(person):
  return person['name']
people=[{"name":"aman","age":21},{"name":"rahul","age":34}]
list(map(get_name,people))
```
