```python
num=int(input("Enter the number:"))
if num%2==0:
  print("The number is even")
else:
  print("The number is odd")
```

```python
def odd_or_even(num):
  if num%2==0:
    print("The number is even")
  else:
    print("The number is odd")

odd_or_even(67) ## callin
```

```python
def add(a,b):
  return a+b
result=add(2,6)
print(result)
```

```python
def greet(name):
  print(f"Hello {name} how are you")

greet("aman")
```

```python
def greet(name="prashant"):  ## by default it will consider prashant if u dont give in greet()
  print(f"Hello {name} how are you")

greet()
```

```python
## variable length arguments
## positional and keyword arguments
def print_numbers(*arggs):  ## gere arggs has values as many parameters
  for i in arggs:
    print(i)
print_numbers(1,2,3,5,6,9,"aman")
```

```python
## keyword arguments
def print_numbers(**kwarggs):
  for key,value in kwarggs.items():
    print(f"{key}:{value}")
print_numbers(name="aman",age=21,roll=35)
```

```python
def print_numbers(*arggs,**kwarggs):
  for i in arggs:
    print(f"positional arguments:{i}")

  for key,value in kwarggs.items():
    print(f"{key}:{value}")

print_numbers(1,2,3,4,5,"rahul",name="aman",age=21,roll=56)
```
