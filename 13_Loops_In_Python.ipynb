```python
## for loop
```

```python
for i in range(5):
  print(i)
```

```python
for i in range(1,6):
  print(i)
```

```python
for i in range(1,10,2):## step size is 2
  print(i)
```

```python
for i in range(10,1,-1): ## except last one ie 1
  print(i)
```

```python
str="aman kumar"  ## as string is a collection of characters
for i in str:
  print(i)
```

```python
## while loop
## The while loop continues to execute if the condition is True
count=0
while count<5:
  print(count)
  count=count+1
```

```python
count=0
while count%2==0:
  print(count)
  count=count+1
```

```python
## Loop control statements
## the  break statements exits the loop permanently
for i in range(10):
  if i==5:
    break
  print(i) ## here print is out of if
```

```python
## the continue statement skips the current iteration  and continues with the next
for i in range(10):
  if i%2==0:
    continue
  print(i)
```

```python
## pass
## The pass statement is a null opearation; it does nothing
for i in range(10):
  if i==3:
    pass
  print(i)
```

```python
## nested loops
## loops inside a loop
for i in range(3):        ## for every i j will run  three  times
  for j in range(3):
    print(f"i:{i} and j:{j}")
```

```python
## to calculate the sum of first N natural number using for loop

num=int(input("Enter the number of natural number:"))
sum=0
for i in range(1,num+1):
  sum=sum+i
print(sum)
```

```python
## using while loop to calculate the sum of first N natural number
N=10
sum=0
count=1
while count<=N:
  sum=sum+count
  count=count+1
print("The sum is:",sum)
```

```python
## Example- Prime numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
