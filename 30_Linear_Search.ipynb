```python


def Linear_search(arr,target):
  size=len(arr)
  for index in range(0,size):
    if arr[index]==target:
     return index
  return -1

my_list=[1,2,3,4,5,78,34,23,62]
Target=78
Result=Linear_search(my_list,Target)
print(Result)
```
